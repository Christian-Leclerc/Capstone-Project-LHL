# Project Workflow

```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np

tasks = [
    {'name': 'Exploring data', 'start': 0, 'duration': 9, 'completion': 100},
    {'name': 'Training model', 'start': 8, 'duration': 3, 'completion': 100},
    # NO TIME --> {'name': 'Testing model', 'start': 8, 'duration': 3, 'completion': 0},
    {'name': 'Dashboard/Presentation', 'start': 11, 'duration': 3, 'completion': 100},
    {'name': 'Dry-Run', 'start': 14, 'duration': 1, 'completion': 100},
    {'name': 'Demo Day', 'start': 15, 'duration': 1, 'completion': 0}
]
current_day = 14

# Calculate total progression
total_duration = sum(task['duration'] for task in tasks)
total_progression = sum(task['duration'] * task['completion'] for task in tasks) / total_duration

# Create colormap
cmap = plt.cm.RdYlGn

plt.figure(figsize=(10, 5))

for i, task in enumerate(tasks):
    if task['start'] > current_day:
        color = 'grey'  # Not started yet
    else:
        color = cmap(task['completion'] / 100.0)
        
    plt.barh(task['name'], task['duration'], left=task['start'], color=color)
    plt.text(task['start'] + task['duration'] + 0.1, i, f"{task['completion']}%")

plt.axvline(x=current_day, color='r', linestyle='--', label=f'Current Day - Total Progress: {total_progression:.2f}%')
plt.xlabel('Days')
plt.ylabel('Tasks')
plt.legend()
plt.gca().invert_yaxis()
```
